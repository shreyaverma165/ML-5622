```python
from sklearn import preprocessing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction import DictVectorizer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import FunctionTransformer, Normalizer
from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.linear_model import SGDClassifier
from csv import DictReader, DictWriter
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize as WordTokenizer

def word_tokenizer(data, column):
    token=[]
    for item in data[column]:
         token.append(WordTokenizer(item))
    return token
  
def posTaggerFeature(examples):
    featureList = []
    tagger = {'VBD': 0, 'VBN': 0, '"': 0, '"': 0, '!':0, '?':0}
    tokens = nltk.word_tokenize(examples)
    tagged_pos = nltk.pos_tag(tokens)
    for tag in tagged_pos:
        if tag in tagger:
            tagger[tag]+=1
    features = {}
    features.update(tagger)
    return features
    
class PartOfSpeechTransformer(TransformerMixin):
    
    def __init__(self):
        pass
    
    def fit(self, examples):
        return self

    def transform(self,examples):
        return examples.apply(posTaggerFeature)

class DataFrameParser(BaseEstimator, TransformerMixin):
    def __init__(self, dataFrame):
        self.dataFrame = dataFrame
    
    def fit(self, examples):
        return self

    def transform(self, examples):
        return examples[self.dataFrame]


class XYZTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        self.feature_names = []
    
    def fit(self, examples):
        return self
    
    def transform(self, examples):
        
        import numpy as np 
        from scipy.sparse import csr_matrix
                 
        # Initiaize matrix 
        X = np.zeros((len(examples), 1))
        
        # Loop over examples and count letters 
        for ii, x in enumerate(examples):
            X[ii, :] = np.array([len(x)])
        return csr_matrix(X)
    
    def get_feature_names(self):
        return self.feature_names
```

```python
from sklearn.utils import shuffle
dfTrain = pd.read_csv("../data/spoilers/train.csv")
class FeatEngr:
    
    def __init__(self):
#         from sklearn.feature_extraction.text import CountVectorizer
        
#         self.feature_union = CountVectorizer()
        self.vectorizer = FeatureUnion([
            ('Part of Speech',Pipeline([('Sentence', DataFrameParser('sentence')),('POS',PartOfSpeechTransformer()),('Dict', DictVectorizer())])),
            ('TF-IDF Bag of Words', Pipeline([('Sentence', DataFrameParser('sentence')),('tf-idf', TfidfVectorizer(analyzer='word',ngram_range=(1,2), lowercase=True, norm='l2',stop_words='english'))])),
#           ('Bag Of Words', Pipeline([('Sentence', DataFrameParser('sentence')),('BOW', CountVectorizer())])),
#             ('word Count', Pipeline([('Sentence', DataFrameParser('sentence')),('Count', XYZTransformer())])),
            ('Trope count', Pipeline([('Sentence', DataFrameParser('trope')),('Tropes', TfidfVectorizer())]))
        ])
        
    def build_train_features(self, examples):
        """
        Method to take in training text features and do further feature engineering 
        Most of the work in this homework will go here, or in similar functions  
        :param examples: currently just a list of forum posts  
        """
        return self.vectorizer.fit_transform(examples)

    def get_test_features(self, examples):
        """
        Method to take in test text features and transform the same way as train features 
        :param examples: currently just a list of forum posts  
        """
        return self.vectorizer.transform(examples)

    def show_top10(self,examples):
        """
        prints the top 10 features for the positive class and the 
        top 10 features for the negative class. 
        """
#         feature_names = np.asarray(self.feature_union.get_feature_names())
        bagofwords = TfidfVectorizer(analyzer='word',ngram_range=(1,2), lowercase=True, norm='l2',stop_words='english')
        Bow = bagofwords.fit_transform(examples[0])
        feature_bow = bagofwords.get_feature_names()
        
        tropecount = TfidfVectorizer()
        Trope = tropecount.fit_transform(examples[1])
        feature_trope = tropecount.get_feature_names()
        
        sentencelength = XYZTransformer()
        Length = sentencelength.fit_transform(examples[0])
        feature_length = sentencelength.get_feature_names()
        
        feature_names = np.asarray(feature_bow+feature_trope+feature_length)
        print("feature names:", feature_names)
        top10 = np.argsort(self.logreg.coef_[0])[-10:]
        bottom10 = np.argsort(self.logreg.coef_[0])[:10]
        print("Pos: %s" % " ".join(feature_names[top10]))
        print("Neg: %s" % " ".join(feature_names[bottom10]))
                
    def train_model(self, random_state=1234):
        """
        Method to read in training data from file, and 
        train Logistic Regression classifier. 
        
        :param random_state: seed for random number generator 
        """
        
        from sklearn.linear_model import LogisticRegression 
        
        # load data 
        self.dfTrain = pd.read_csv("../data/spoilers/train.csv")
        shuffle(self.dfTrain)
        #Doing 90/10 Cross Validation 
        #Train Ratio = 0.9
        #Test Ratio = 0.1
        self.dfTrain = self.dfTrain[0:int(0.8*round(len(self.dfTrain)))]
        
        # get training features and labels 
        from sklearn.linear_model import lasso_path
        from sklearn.feature_selection import SelectFromModel
        from sklearn.linear_model import LassoCV
        
        self.X_train = self.build_train_features(self.dfTrain)
        self.y_train = np.array(self.dfTrain["spoiler"], dtype=int)
        
#         _, coef_path, _ = lasso_path(self.X_train, self.y_train)
#         print(coef_path)
        
#         clf = LassoCV()

#         # Set a minimum threshold of 0.25
#         model = SelectFromModel(clf, threshold=0.25)
#         model.fit(self.X_train, self.y_train)
#         n_features = model.transform(self.X_train).shape[1]
#         print(n_features)
        
        # train logistic regression model.  !!You MAY NOT CHANGE THIS!! 
        self.logreg = LogisticRegression(random_state=random_state)
        self.logreg.fit(self.X_train, self.y_train)
    
    def classification_visualizer(self):
        """
        Method to incoporate various classification and text model visualizers 
        and predict the accuracy of your train model.
        """ 
        from sklearn.linear_model import lasso_path
        
        #Classification Report: Presenting classifier as a heatmap
        cross_validation = self.dfTrain[int(0.8*round(len(self.dfTrain))):]
        self.X_validate = self.get_test_features(cross_validation)
        self.y_validate = np.array(cross_validation["spoiler"], dtype=int)
        y_pred = list(self.logreg.predict(self.X_validate))
        target_labels = ["spoiler", "not spoiler"]
        
        visualize = classification_report(self.y_validate, y_pred, target_names=target_labels)
        print("Classification Report: \n", visualize)
        
        print("y_pred has type ", type(y_pred))
        print("X_validate has type ", type(self.X_validate))
        print("y_validate has type ", type(self.y_validate))
        print("X_validate has shape ", self.X_validate.shape)
        print("y_validate has shape", self.y_validate.shape)
#         print(self.X_validate.todense())
        
        #Confusion Matrix Analysis
        matrix_visualize = confusion_matrix(self.y_validate, y_pred)
        print("\n Confusion Matrix: \n", matrix_visualize)

    def model_predict(self):
        """
        Method to read in test data from file, make predictions
        using trained model, and dump results to file 
        """
        
        # read in test data 
        dfTest  = pd.read_csv("../data/spoilers/test.csv")
        
        # featurize test data 
        self.X_test = self.get_test_features(dfTest)
        
        # make predictions on test data 
        pred = self.logreg.predict(self.X_test)
        
#         misclassified = pd.Series(np.array(self.X_test['spoiler']) != y)
#         print(self.mytest[misclassified.values][['spoiler', 'sentence']])
        
        # dump predictions to file for submission to Kaggle  
        pd.DataFrame({"spoiler": np.array(pred, dtype=bool)}).to_csv("prediction.csv", index=True, index_label="Id")
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

dfTrain = pd.read_csv("../data/spoilers/train.csv")

spoilers_length = []
non_spoilers_length = []

for index,row in dfTrain.iterrows():
    if row['spoiler'] == True:
        spoilers_length.append(len(row['sentence']))
    else:
        non_spoilers_length.append(len(row['sentence']))

plt.bar(range(0,len(spoilers_length)),spoilers_length,label='Spoiler Count')
plt.bar(range(0,len(non_spoilers_length)),non_spoilers_length,label='Non Spoiler Count')
plt.title('Feature 2: Spoiler Analysis based on Length')
plt.xlabel('ID')
plt.ylabel('length of sentence')
plt.legend(loc = 'best')
plt.show()
```
