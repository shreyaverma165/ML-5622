### [15 points] Problem 1: 
***

Consider the class C of concepts defined by triangles with **distinct** vertices of the form $(i, j)$ where $i$ and $j$ are integers in the interval $[0,99]$. A concept c labels points on the interior and boundary of the associated triangle as positive and points exterior to the triangle as negative.

**Note**: To make life easier, we'll allow degenerate triangles in $C$. That is, triangles where the vertices are collinear. The following image depicts an example of a nondegenerate and a degenerate triangle.

<img src="figs/triangles.png" width=400 height=50>  

**Part A**: Suppose we have an algorithm that produces a consistent $h$ from the hypothesis class $H = C$. Give a bound on the number of training examples sufficient to assure that for any target concept $c$ in $C$, our algorithm will, with probability $1-\delta$, output a hypothesis $h$ with generalization error at most $\epsilon$.

**Answer:**

As stated by the problem, we have $\epsilon = 0.15$ : the error of the hypothesis $h_{i}$. The probability of hypothesis
h being outputted with error $\epsilon$ is the confidence of the hypothesis which is given by the equation $1-\delta$. In
this particular problem, $1-\delta = 0.95$. This means $\delta = 0.05$.

This problem has a finite, consistent hypothesis class $H$. It is finite because each hypothesis $h_{i}$ is a
triangle with 3 distinct vertices on the interval $[0,99]$: there is a maximum number of triangles. It is consistent
because each $h_{i}$ determines if a given a training example $x_{i}$ is inside or outside of its boundaries. This
is aligned with the concept $c$ of labeling each point positive or negative depending on whether that point
is interior and exterior to the triangle boundary. 

There is an algorithm A that exists that, given a point $x_{i}$ and 3 vertices that comprise a triangle, will tell if that point is inside or outside of the triangle boundary. Because this problem has a finite, consistent hypothesis class we can use the following equation to find the bound on training examples $m$ to ensure each hypothesis has a confidence of $95$% and error of $0.15$.
    
$$ m ≥  \frac{1}{ϵ} (ln ⁡|H|+ln⁡ \frac{1}{δ})$$
-----------------------------------------------------------------------------------------------------------------------

**Part B**: Based on your bound in **Part A**, determine the minimum number of training examples necessary such that for any target concept $c$ in $C$, our algorithm will, with probability $0.95$, output a hypothesis $h$ with generalization error at most $0.15$.  

**Answer:**

$|H|$ is the number of hypothesis which is equal to the number of triangles that can be found in the problem. The number of triangles that can be found in a given interval is the total number of combinations of the total number of points in that interval made up of $3$ vertices $C (n,3)$. $n$ is $100$ points in the case when the interval for both x and y is $[0,99]$. This means there are $C (100,3)$ * $ C (100,3)$ = $26146890000$ possible triangles. This means $|H| = 26146890000$. 

Plugging $|H|, \epsilon, \delta$ in the above equation, we get

$$ m ≥  \frac{1}{0.15} (ln ⁡(26146890000)+ln⁡ \frac{1}{0.05}) =  179.884 $$
---------------------------------------------------------------------------------------------------------------
Therefore, total $180$ training examples.

    From the above figure, the upper bound of the VCdim(H) should be 2 as well. 
    As depicted in fifth plot, there is no way for a hyperplane to classify all labelings of two points when the point
    between the other two points is a different label. This is true for all configurations of the points.

In the case of 2 dimensions, the hyperplane (decision boundary) can be flattened into a line represented by the equation $y = wx$, where $w$ is the slope of the line. The hypothesis class is then the set of all these equations for a line that classify m points correctly. 

Let us label points positive if they can be intersected by a line with slope $r ≥ w$ and points negative if they can be intersected by a line with slope r < w. This gives us the function for each hypothesis: 
                        $ h(p = (x, y)) = +1$ if $r = \frac{y}{x} ≥ w $ and 
                        $ h(p = (x, y)) = -1$ if $r = \frac{y}{x} < w $
                        

**Proof:**

    We want to prove that for all configurations of the points there exists a labelling that cannot be shattered
    by the hypothesis class H.
    Let p1, p2, p3 be points such that their y and x coordinates are ordered as (x1, y1) , (x2, y2) , (x3; y3). 
    Let the class labels for each point be set to p1 = +1; p2 = -1; p3 = +1. 
    Let w be the radius of the hypothesis h. 
    Since p1 = +1, the radius of p1, r1, must be ≥ w. 
    Since p3 = +1, the radius of p3, r3, must be < w. 
    This would mean, based on the ordering of the coordinates above, that the radius of p2, r2, must also be ≥ w.
    But this contradicts the original labelling of p2 as -1 which would mean r2 < w. 
    This means that there is no hypothesis h that can shatter a set of 3 points. 
    Therefore, VCdim < 3 or VCdim = 2.

```python
###############################################################################################
# Write some code that randomly generates a concept rectangle c.
###############################################################################################

import numpy as np
from random import uniform

def generate_rectangle_C(width, height):
    rectangle_width = uniform(width,height)
    rectangle_height = uniform(width,height)
    vertex_a = (0, 0); vertex_b = (rectangle_width, 0)
    vertex_c = (rectangle_width, rectangle_height); vertex_d = (0, rectangle_height)
    rectangle_dimensions = ['A', 'B', 'C', 'D', 'width', 'height']
    rectangle_dimensions = [vertex_a, vertex_b, vertex_c, vertex_d, rectangle_width, rectangle_height]
    return rectangle_dimensions

def return_point_labels(Xval,Yval,limit):
    if ((limit[0][0] <= Xval <= limit[1][0]) and limit[0][1] <= Yval <= limit[3][1]): 
        return 1
    else:
        return -1
```

```python
def return_smallest_rectangle(feature_vector,labels, class_bounds):
    for feature, label in zip(feature_vector, labels):
        if label == 1:
            hypothesis_lower_x = max(class_bounds[0][0], feature[0]-1)
            hypothesis_lower_y = max(class_bounds[0][1], feature[1]-1)
            hypothesis_width = 2
            if hypothesis_lower_x+hypothesis_width > class_bounds[1][0]:
                hypothesis_width = class_bounds[1][0] - (hypothesis_lower_x+width)
            hypothesis_height = 2
            if hypothesis_lower_y+hypothesis_height > class_bounds[2][1]:
                hypothesis_height = (class_bounds[2][1] - hypothesis_lower_y+height)
            return (hypothesis_lower_x, hypothesis_lower_y), hypothesis_width, hypothesis_height
        else:
            continue

rectangle_bounds = generate_rectangle_C(0,100)
feature_vector = [(43,110), (12,23)]
labels = [return_point_labels(f1,f2,rectangle_bounds) for f1,f2 in feature_vector]
print(labels)
            
fig2 = plt.figure(figsize=[8,8])
ax2= fig2.add_subplot(1,1,1) 
ax2.add_patch(patches.Rectangle((0,0),0,100,linewidth=1,edgecolor='g',facecolor='none'))
for sample,label in zip(feature_vector, labels):
    if label == 1:
        ax2.scatter(sample[0],sample[1], color = '#a76c6e')
    else:
        ax2.scatter(sample[0],sample[1], color = 'steelblue')
lower_xy, h_width, h_height = return_smallest_rectangle(feature_vector,labels,rectangle_bounds)
ax2.add_patch(patches.Rectangle(lower_xy, h_width, h_height,edgecolor='r',fill= False))
ax2.plot()
```

```python
def generate_hypothesis_H(Xval,Yval):
    left_bound = min(Xval); right_bound = max(Xval); bottom_bound = min(Yval); upper_bound = max(Yval)
    hypothesis_vertex_A = (left_bound, bottom_bound)
    hypothesis_vertex_B = (right_bound, bottom_bound)
    hypothesis_vertex_C = (right_bound,upper_bound)
    hypothesis_vertex_D = (left_bound,upper_bound)
    hypothesis_class_limits = ['A', 'B', 'C', 'D']
    hypothesis_class_limits = [hypothesis_vertex_A, hypothesis_vertex_B, 
                               hypothesis_vertex_C, hypothesis_vertex_D]
    return hypothesis_class_limits
```

```python
def return_samples(size):
    x_indices = np.random.choice(np.arange(-40,100), size)
    y_indices = np.random.choice(np.arange(-40,100), size)
    labels = [return_point_labels(x,y,rectangle_bounds) for x,y in zip(x_indices,y_indices)]
    return [(x,y) for x,y in zip(x_indices,y_indices)], labels

def return_error(test, predicted):
    increment = 0
    for num,(t,p) in enumerate(zip(test,predicted)):
        if t==p:
            increment+=1
    return round(1-(increment/float(len(test))),5)
```

```python
size = 1500
feature_set,label_set = return_samples(size)
X_example_train_set = feature_set[:-1000]; y_example_train_set = label_set[:-1000]
fig3 = plt.figure(figsize=[8,8])
ax3= fig3.add_subplot(1,1,1)
ax3.add_patch(patches.Rectangle(rectangle_bounds[0],rectangle_bounds[4],rectangle_bounds[5],edgecolor='b',fill=False))
for index,label in zip(X_example_train_set, y_example_train_set):
    if label == 1:
        ax3.scatter(index[0],index[1], color = '#a76c6e', s=30)
    else:
        ax3.scatter(index[0],index[1], color = 'steelblue', s=30)

x_coordinate = []; y_coordinate = []
for index,label in zip(X_example_train_set, y_example_train_set):
    if label == 1:
        x_coordinate.append(index[0])
        y_coordinate.append(index[1])

H_limit = generate_hypothesis_H(x_coordinate, y_coordinate)
ax3.add_patch(patches.Rectangle(H_limit[0], H_limit[1][0] - H_limit[1][0], H_limit[3][1] - H_limit[1][1],
                                linewidth = 2, edgecolor='g',fill=False))
ax3.plot()
```

```python
####################################################################################################################
# Write some code that computes approximate generalization errors for  100  independent concepts  c  
# and associated training sets of size  m , and returns the worst-case generalization error at the confidence level  1−δ. 
# One way to do this in the case that say  δ=0.05 , is to report the  95th  percentile of the 100 samples of the 
# generalization error. We can then say that, in our simulation,  100(1−δ)%  of our observed generalization errors 
# were less than our computed value. 
# (Bonus: If your code is efficient, try increasing the number of runs in the simulation to  500 . 
#  This should give you a better approximation of the generalization error.)
####################################################################################################################

concepts = 100

def calculate_percentile_generalization_error(concepts, size):
    XY = (0,99)
    err = []
    for val in range(100):
        C_limit = generate_rectangle_C(XY[0],XY[1])
        feature_set,label_set = return_samples(size)
        X_example_train_set = feature_set[:-1000]; y_example_train_set = label_set[:-1000]
        X_example_validation_set = feature_set[-1000:]; y_example_validation_set = label_set[-1000:]
        x_coordinate = []; y_coordinate = []
        for index,label in zip(X_example_train_set, y_example_train_set):
            if label == 1:
                x_coordinate.append(index[0])
                y_coordinate.append(index[1])

        H_limit = generate_hypothesis_H(x_coordinate, y_coordinate)
        
        test_c = [return_point_labels(s1,s2,rectangle_bounds) for s1,s2 in X_example_validation_set]
        predicted_c = [return_point_labels(s1,s2,H_limit) for s1,s2 in X_example_validation_set]
        err.append(return_error(test_c, predicted_c))
    return round(np.percentile(err, 95),7) 

print("For 100 simulations: {0}".format(calculate_percentile_generalization_error(concepts, size)))
print("For 500 simulations: {0}".format(calculate_percentile_generalization_error(500, size)))
```

```python
def unif(size, bounds):
    XY=(0,100)
    Xx_m = np.random.choice(np.arange(XY[0],XY[1]), size)
    Yy_m = np.random.choice(np.arange(XY[0],XY[1]), size)
    C_label = [return_point_labels(x1,y1,bounds) for x1,y1 in zip(Xx_m,Yy_m)]
    return [(x1,y1) for x1,y1 in zip(Xx_m,Yy_m)], C_label

def calculate_generalization_error_m(concepts, size):
    XY = (0,99)
    error = []

    for index in range(100):
        C_limit = generate_rectangle_C(XY[0],XY[1])
        feature_set,label_set = unif(size, C_limit)
        X_example_train_set = feature_set[:int(round(0.8*len(feature_set)))]
        y_example_train_set = label_set[:int(round(0.8*len(label_set)))]
        X_example_validation_set = feature_set[int(round(0.8*len(feature_set))):]
        y_example_validation_set = label_set[int(round(0.8*len(label_set))):]
        x_coordinate = []; y_coordinate = []
        for index,label in zip(X_example_train_set, y_example_train_set):
            if label == 1:
                x_coordinate.append(index[0])
                y_coordinate.append(index[1])
        
        test_m = [return_point_labels(s1,s2,rectangle_bounds) for s1,s2 in X_example_validation_set] 
        if len(x_coordinate) > 0:
            H_limit = generate_hypothesis_H(x_coordinate, y_coordinate)   
            pred_m = [return_point_labels(s1,s2,H_limit) for s1,s2 in X_example_validation_set]
        else:
            H_limit = [(0, 0), (0, 0), (0, 0), (0, 0)]
            pred_m = [-1]*len(test_m)

        error_test_predicted = return_error(test_m, pred_m)
        error.append(error_test_predicted)
    return round(np.percentile(error, 95),3)

num = 500
m= [250, 500, 1000, 1250, 1500]
generalization_error = []
for size in m:
    err_1 = calculate_generalization_error_m(num, size)
    generalization_error.append(err_1)
print(generalization_error)
```

```python
plt.loglog(m, generalization_error, basex=np.e, basey=np.e)
plt.xlabel("Number of training samples")
plt.ylabel("Generalization error with confidence = 0.05")
plt.xticks(m, m)
plt.title("Generalization error vs number of samples")
plt.show()
```

**Theoretical PAC bound Calculation**
According to the question, the value of $\delta$ is 0.05 and the value of $\epsilon$ can be taken to be 0.15.
Therefore, from the formula:

$$
m > \frac{4}{\epsilon}\log\frac{4}{\delta}
$$
--------------------------------------------------------------------------------------------------------
which gives:

$$
m > \frac{4}{0.15}\log\frac{4}{0.05}
$$
---------------------------------------------------------------------------------------------------------
m > 50.74

m ~= 51
