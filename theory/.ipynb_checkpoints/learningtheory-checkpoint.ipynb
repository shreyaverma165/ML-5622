### [15 points] Problem 1: 
***

Consider the class C of concepts defined by triangles with **distinct** vertices of the form $(i, j)$ where $i$ and $j$ are integers in the interval $[0,99]$. A concept c labels points on the interior and boundary of the associated triangle as positive and points exterior to the triangle as negative.

**Note**: To make life easier, we'll allow degenerate triangles in $C$. That is, triangles where the vertices are collinear. The following image depicts an example of a nondegenerate and a degenerate triangle.

<img src="figs/triangles.png" width=400 height=50>  

**Part A**: Suppose we have an algorithm that produces a consistent $h$ from the hypothesis class $H = C$. Give a bound on the number of training examples sufficient to assure that for any target concept $c$ in $C$, our algorithm will, with probability $1-\delta$, output a hypothesis $h$ with generalization error at most $\epsilon$.

**Ans:** The bound on the number of training examples for a **consistent, finite hypothesis class $H = C$** is given by: 
### $$m ≥ \frac{1}{ϵ} (ln ⁡|H|+ln⁡ \frac{1}{δ})$$

**Part B**: Based on your bound in **Part A**, determine the minimum number of training examples necessary such that for any target concept $c$ in $C$, our algorithm will, with probability $0.95$, output a hypothesis $h$ with generalization error at most $0.15$.  

**Ans:** Given generalization error, $ϵ = 0.15$
<br> Confidence, $1-δ = 0.95 => δ = 0.05$
<br> Since the vertices i and j are in the interval $[0,99]$, there a a total of $100*100 = 10000$ combinations of each vertex of the triangle. Given that, the total number of triangles possible is: 
\begin{equation*} {10000 \choose 3} = 166616670000 \end{equation*}
<br> $|H| = 166616670000$.
Repacing in the above bound, $$m ≥ \frac{1}{0.15} (ln(166616670000)+ln⁡ \frac{1}{0.05})$$
$m ≥ 192.23$.
<br> Hence, the **mimimum number of training examples required is 193**.

As seen from the fouth plot above, there exists no circle centered around the origin such that the circle does not also include the negative sample.
<br> **Proof:** 
<br> Consider two points such that their distance from the center (or radii of the circle formed by the point) are r1 and r2 . Let their distances be such that r1<r2.
<br> Consider the labels r$1 = -1$ and $r2 = +1$.
<br> Let r be the radius of the hypothesis h. 
    Since r1 = -1, the radius  r1, must be > r. 
    Since r2 = +1, the radius r2 <= r. 
    According to above logic, r2 < r1. But the initial asumption made was that r2 > r1. 
    Hence, there is no hypothesis circle h that can shatter a set of 2 points or  $VC~dim < 2$
<br>**Therefore, $VC~dim = 1$.**

In the in-class notebook associated with PAC Learnability, we proved a PAC bound for the class of concepts $C$ comprised of axis-aligned rectangles living in $\mathbb{R}^2$ of the form $(a \leq x \leq b) \wedge (c \leq y \leq d)$ where $a, b, c, d$ are real numbers. Specifically, we proved that with probability $1-\delta$, any consistent learner could learn a hypothesis $h$ in $H = C$ with generalization error less than $\epsilon$ provided that the number of training examples satisfied

```python
import matplotlib.pylab as plt
%matplotlib inline

from random import randint, sample

import matplotlib.patches as mpatches

def generate_concept_class(lower_limit, upper_limit):
    ## Assume rectangles only in the first quadrant for simplicity
    ## Generate a concept class with width and height in the range [0,99] inclusive
    x_left_bottom = 0 #Choose a random integer from 0 to 15 as the left bottom of the rectangle
    y_left_bottom = 0
    concept_width = randint(lower_limit,upper_limit) #Width (x in the features)
    concept_height = randint(lower_limit,upper_limit) #Height (y in the features)
    A = (0, 0)
    B = (concept_width, 0)
    C = (concept_width, concept_height)
    D = (0, concept_height)
    concept_class_bounds = {'A': A, 'B': B, 'C': C, 'D': D, 'width': concept_width, 'height': concept_height}
    return concept_class_bounds
```

```python
## Assume that the features are the x and y coordinated of the point.
def assign_label(x_in,y_in,bounds):
    """If the point is within or on the bounds of the rectangle label = 1"""
    if ((bounds['A'][0]<=x_in<=bounds['B'][0]) \
        and bounds['A'][1]<=y_in<=bounds['D'][1]): 
        return 1
    else:
        return -1
```

```python
def get_smallest_rectangle(feature_vector,labels, class_bounds):
    for feature, label in zip(feature_vector, labels):
        if label == 1:
            ## Create the smallest possible rectangle about the positive sample.
            hypothesis_lower_x = max(class_bounds['A'][0], feature[0]-1)
            hypothesis_lower_y = max(class_bounds['A'][1], feature[1]-1)
            hypothesis_width = 2
            if hypothesis_lower_x+hypothesis_width > class_bounds['B'][0]:
                hypothesis_width = class_bounds['B'][0] - (hypothesis_lower_x+width)
            hypothesis_height = 2
            if hypothesis_lower_y+hypothesis_height > class_bounds['C'][1]:
                hypothesis_height = (class_bounds['C'][1] - hypothesis_lower_y+height)
            return (hypothesis_lower_x, hypothesis_lower_y), hypothesis_width, hypothesis_height
        else:
            continue
```

```python
def generate_hypothesis_class(x_coords,y_coords):
    left_lim = min(x_coords)
    right_lim = max(x_coords)
    bottom_lim = min(y_coords)
    upper_lim = max(y_coords)
    hyp_A = (left_lim, bottom_lim)
    hyp_B = (right_lim, bottom_lim)
    hyp_C = (right_lim,upper_lim)
    hyp_D = (left_lim,upper_lim)
    hypothesis_class_bounds = {'A': hyp_A, 'B': hyp_B, 'C': hyp_C, 'D':hyp_D}
    return hypothesis_class_bounds
```

```python
import numpy as np

def generate_samples(size):
    x_samples = np.random.choice(np.arange(-20,100), size)
    y_samples = np.random.choice(np.arange(-20,100), size)
    concept_class_labels = [assign_label(x,y,concept_class_bounds) for x,y in zip(x_samples,y_samples)]
    return [(x,y) for x,y in zip(x_samples,y_samples)], concept_class_labels
```

```python
def get_err(test_samples, pred_samples):
    ## Get the generalization error given the test labels and predicted labels
    correct = 0
    for num,(t,p) in enumerate(zip(test_samples,pred_samples)):
        if t==p:
            correct+=1
    accuracy = correct/float(len(test_samples))
    return round(1-accuracy,7)
```

**Part B**: Use your code to estimate the generalization error with confidence parameter $\delta=0.05$ for training sets of size $m$ where $m = 250, 500, 1000, 1250,$ and $1500$ and the data are comprised of points $(x,y)$ where the $x$- and $y$-values are sampled from the continuous uniform distribution $\textrm{unif}(0,100)$. Make a **log-log** plot with $m$ on the horizontal axes and $\epsilon$ on the vertical axis.  Additionally, overlay the theoretical PAC bound on your graph and discuss your results. 

```python
def unif(size, bounds):
    pool=(0,100)
    x_samples = np.random.choice(np.arange(pool[0],pool[1]), size)
    y_samples = np.random.choice(np.arange(pool[0],pool[1]), size)
    cc_labels = [assign_label(x,y,bounds) for x,y in zip(x_samples,y_samples)]
    return [(x,y) for x,y in zip(x_samples,y_samples)], cc_labels
```

```python
def gen_err_variable_size(num_simulations, size):
    """Vary size to get generaliation errors. Take 80% of data as train and rest as validation set"""
    xy_range = (0,99)
    gen_errors_m = []

    for index in range(100):
        c_class_bounds = generate_concept_class(xy_range[0],xy_range[1])
        features,given_labels = unif(size, c_class_bounds)
        ## Split the data to test and cross - validation sets
        X_train = features[:int(round(0.8*len(features)))]
        y_train = given_labels[:-int(round(0.8*len(given_labels)))]

        ## Take the last m samples of the dataset as validation set
        X_valid = features[int(round(0.8*len(features))):]
        y_valid = given_labels[int(round(0.8*len(given_labels))):]

        ## Get the x and y coordinates of positive samples in the training set
        x_pos = []
        y_pos = []
        for sample,label in zip(X_train, y_train):
            if label == 1:
                x_pos.append(sample[0])
                y_pos.append(sample[1])

        y_test_m = [assign_label(f1,f2,concept_class_bounds) for f1,f2 in X_valid] # test labels
        if len(x_pos) > 0:
            h_class = generate_hypothesis_class(x_pos, y_pos)    ## Prediction on the validation set
            y_pred_m = [assign_label(f1,f2,h_class) for f1,f2 in X_valid]
        else:
            h_class = {'A': (0, 0), 'B': (0, 0), 'C': (0, 0), 'D': (0, 0)}
            y_pred_m = [-1]*len(y_test_m)

        error_pred = get_err(y_test_m, y_pred_m)
        gen_errors_m.append(error_pred)
    return round(np.percentile(gen_errors_m, 95),3)
```

```python
## Get the generalization errors for a list of variable sizes for 500 simulations
num_simulations = 500
m= [250, 500, 1000,1250]
generalization_err = []
for size in m:
    error_temp = gen_err_variable_size(num_simulations, size)
    generalization_err.append(error_temp)
print(generalization_err)
```

```python
plt.bar(m, generalization_err)
plt.xlabel("Number of training samples")
plt.ylabel("Generalization error with confidence = 0.05")
plt.xticks(m, m)
plt.title("Generalization error vs number of samples")
plt.show()
```
